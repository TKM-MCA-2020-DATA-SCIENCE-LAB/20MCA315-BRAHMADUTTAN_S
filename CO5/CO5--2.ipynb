```python
from bs4 import BeautifulSoup
import csv
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
webpage = requests.get("https://www.flipkart.com/search?q=poco&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
```

```python
webpage
```

```python
soup = BeautifulSoup(webpage.content,"html.parser")
```

```python
print(soup.prettify())
```

```python
poco_models = []
for i in range(0,len(names)):
  poco_models.append(names[i].get_text())
```
